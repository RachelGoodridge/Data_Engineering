# Project Proposal

### Question/Need:
1. What is the end service that your project will provide? What is the purpose of the system you plan to build?
    - The end service is an interactive website application that is meant to inform the public about the rates of COVID diagnoses and vaccinations across the United States.
2. Who is your client and how will that client benefits from your end service?
    - The client is any health organization who is looking to discover where COVID-19 outreach is most needed in the US (i.e. places with low vaccination rates and high disease rates) and will focus their efforts on educating / promoting the vaccine in those areas.
3. What does your end-to-end data pipeline look like? **You should submit a preliminary sketch of your pipeline in the form of a diagram or workflow list with your proposal.** For more directions and examples see the Data Pipeline and Project Workflow pages.

### Data Description:
1. What dataset(s) do you plan to use, and how will you obtain the data? Please include a link! (The link can be to the dataset you’re downloading, the site you’re scraping, etc.)
    - The first dataset on COVID cases is from [worldometer](https://www.worldometers.info/coronavirus/country/us/) and can also be downloaded through [kaggle](https://www.kaggle.com/imdevskp/corona-virus-report). It currently has 628k rows and 14 columns.
    - The second dataset on COVID vaccines is from the [CDC](https://covid.cdc.gov/covid-data-tracker/#vaccinations) and can also be downloaded through [github](https://github.com/owid/covid-19-data/tree/master/public/data/vaccinations). It currently has 25k rows and 16 columns.
2. Do you plan to be able to automatically pull in new data at a regular cadence (e.g with Airflow or a cron job)?
    - Yes, these datasets are updated daily.
3. What is an individual sample/unit of analysis in this project? In other words, what does one row or observation of the data represent?
    - In the COVID cases dataset, a row represents a county in the US on a particular day.
    - In the COVID vaccines dataset, a row represents a state in the US on a particular day.
4. What characteristics/features do you expect to work with? In other words, what are your columns of interest?
    - In the COVID cases dataset, the columns of interest include the county, state, latitude, longitude, date, confirmed cases, and confirmed deaths.
    - In the COVID vaccines dataset, the columns of interest include the date, state, people vaccinated, people fully vaccinated, percent vaccinated, percent fully vaccinated, boosters, and percent boosted.
5. If modeling, what will you predict as your target?
    - I will be making an automatically updating summary analysis dashboard, not modeling.

### Tools:
1. How do you intend to meet the tools requirements of the project (data storage solution and at least one tool for cloud computing, big data handling, or web deployment)?
    - The data storage tool I am using is SQL.
    - From the category "Web Applications and Model Deployment," I plan to create a Flask web app.
2. Are you planning in advance to need or use additional tools beyond those required?
    - I'll be using Tableau to overlay these data on top of each other in an interactive map format.

### MVP Goal:
1. What would a minimum viable product (MVP) look like for this project?
    - An MVP would be using cleaned data from SQL to produce a summary analysis and visualize this on a map of the United States.
    - Then, deploy this through a Flask web app for users to explore.
    - Later on (post MVP), add more features / interactive components and introduce automatic daily updates.
